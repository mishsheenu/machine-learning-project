```python
# removing outliers using quartile method

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
threshold = 1



```

```python
outliers = ((df < (Q1 - threshold * IQR)) | (df > (Q3 + threshold * IQR))).any(axis=1)
df = df[~outliers]
print("\nNumber of outliers removed for each feature:")
print(outliers.sum(axis=0))
```
